```python
def filter_func(row):
  if row['Attrition']=="Yes" and row['DailyRate']>1372:
    return row['MonthlyIncome']*1.1
  else:
    return row['MonthlyIncome']
```

```python
filter_df=hr_df.copy()
filter_df['MonthlyIncome']=hr_df.apply(filter_func,axis=1)
filter_df.head()
```

```python
filter_df2=filter_df.copy()
```

```python
def filter_advance(row):
  if row['Attrition']=="Yes" and row['DailyRate']>1000 and pd.isna(row['MonthlyIncome'])!=True:
    row['MonthlyIncome']=row['MonthlyIncome']/2
    row['DailyRate']=(row['DailyRate']*10)+100
    return row
  else:
    return row
```

```python
filter_df2=filter_df2.apply(filter_advance,axis=1)
filter_df2.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```
